# Analiza Wariancji

```python
plt.figure(figsize=(20,10))

plt.scatter(range(len(gory)), gory[:,0], c='r')

plt.scatter(range(len(ostance)), ostance[:,0], c='g')
plt.axhline(np.mean(ostance[:,0]) + 2*np.std(ostance[:,0]), c='g')
plt.axhline(np.mean(ostance[:,0]) - 2*np.std(ostance[:,0]), c='g')

plt.scatter(range(len(rowniny)), rowniny[:,0], c='b')

plt.show()
```

```python
print(np.mean(ostance[:,0]) + 2*np.std(ostance[:,0]))
print(np.mean(ostance[:,0]) + 1*np.std(ostance[:,0]))
print(np.mean(ostance[:,0]))
print(np.mean(ostance[:,0]) - 1*np.std(ostance[:,0]))
print(np.mean(ostance[:,0]) - 2*np.std(ostance[:,0]))
print(np.std(ostance[:,0]))
```

```python
plt.plot(np.arange(0,15), st.t.pdf(np.arange(0,15), len(ostance)-1, loc=np.mean(ostance[:,0]), scale=np.std(ostance[:,0])))
st.t.cdf(1.25, len(ostance)-1, loc=np.mean(ostance[:,0]), scale=np.std(ostance[:,0]))
st.t.ppf([0.05,0.95], len(ostance)-1, loc=np.mean(ostance[:,0]), scale=np.std(ostance[:,0]))
print(len(ostance))
```
